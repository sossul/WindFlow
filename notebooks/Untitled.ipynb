```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "../raw_data/2022_viento_dmc.csv"
df = pd.read_csv(url)
df
```

```python
df[['dd_Valor', 'ff_Valor']].plot(kind='line', figsize = (18,18), cmap='coolwarm', subplots = True, fontsize = 10)
plt.show()
```
